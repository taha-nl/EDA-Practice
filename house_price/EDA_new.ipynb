## Missing values

```python
#find_out fratures with null values
features_with_na=[feature for feature in dataset.columns if dataset[feature].isnull().sum()>0]
#the feature and the percentage of missing values
for feat in features_with_na:
    print(feat,np.round(dataset[feat].isnull().mean(),4),'% missing values')
```

### Since we have many missing values we have to find relationship between missing values and Sales Price
let's plot some diagram

```python
for feature in features_with_na:
    data=dataset.copy()
    #let's make a variable that indicates 1 if th observation was missing else 0
    data[feature]=np.where(data[feature].isnull(),1,0)
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.title(feature)
    plt.show()
```

since there an obvious dependency between independent and dependent variable we need to reolace null values with somthing meaningful

## Numercal variables

```python
numerical_features=[feature for feature in dataset.columns if dataset[feature].dtypes !='O']
print('Number of numerical features ',len(numerical_features))
dataset[numerical_features].head()
```

Temporel feature

```python
year_feature=[feature for feature in numerical_features if 'Yr' in feature or 'Year' in feature]
year_feature
```

```python
# Lets analyse the Temporal Datetime Variables
## we willl check whether there is a relation between year the house is sold and the output

dataset.groupby('YrSold')['SalePrice'].median().plot()
plt.ylabel('Median House Price')
plt.xlabel('Year Sold')
plt.title('House Price vs YearSold')
```

```python
for feature in year_feature:
    if feature!='YrSold':
        data=dataset.copy()
        data[feature]=data['YrSold']-data[feature]
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.show()
```

```python
##Numerical variables are usually of 2 type
##1.continous variable and Discrete Variables

discrete_features=[feature for feature in numerical_features if len(dataset[feature].unique())<25 and feature not in year_feature+['Id']]
len(discrete_features)
```

```python
dataset[discrete_features].head()
```

```python
##Lets Find the relationship between them and SalePrice

for feature in discrete_features:
    data=dataset.copy()
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.title(feature)
    plt.show()
```

Continous Variable

```python
continous_feature=[feature for feature in numerical_features if feature not in discrete_features+['Id'] +year_feature]
```

```python
len(continous_feature)
```

```python
for feature in continous_feature:
    data=dataset.copy()
    data[feature].hist(bins=25)
    plt.xlabel(feature)
    plt.ylabel('Count')
    plt.title(feature)
    plt.show()
```

## we will be using logaritmitic transformation

```python

for feature in continous_feature:
    data=dataset.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature])
        data['SelePrice']=np.log(data['SalePrice'])
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.title(feature)
        plt.show()

```

## Outliers

```python
for feature in continous_feature:
    data=dataset.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature])
        data.boxplot(column=feature)
        plt.ylabel(feature)
        plt.title(feature)
        plt.show()
```
